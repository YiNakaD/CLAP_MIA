```python
import json
```

```python
filepath = '../data/LibriSpeech/SPEAKERS.TXT'


def get_speakers(filepath):
    ID = []
    SEX = []
    SUBSET = []
    MINUTES = []
    NAME = []

    with open(filepath, 'r') as f:
        lines = f.readlines()
        for line in lines:
            # print(line)
            lis = str(line).split('|')
            ID.append(lis[0].strip())
            SEX.append(lis[1].strip())
            SUBSET.append(lis[2].strip())
            MINUTES.append(lis[3].strip())
            NAME.append(lis[4].strip())
    return ID, SEX, SUBSET, MINUTES, NAME


speaker_id, sex, _, _, speaker_name = get_speakers(filepath)
print(speaker_name[0])
```

```python
data = {}
for i in range(len(speaker_id)):
    data[speaker_id[i]] = {
        'sex': sex[i],
        'name': speaker_name[i],   
        # chapter_id, reader, _, _, _, chapter_book, chapter_title
    }


print(data)
json_data = json.dumps(data, indent=4)
print(json_data)
# with open('data.json', 'w') as json_file:
#     json.dump(data, json_file, indent=4)
```

```python
filepath = '../data/LibriSpeech/BOOKS.TXT'


def get_books(filepath):
    ID = []
    BOOK = []
    AUTHOR = []

    with open(filepath, 'r') as f:
        lines = f.readlines()
        for line in lines:
            # print(line)
            lis = str(line).split('|')
            # print(lis)
            ID.append(lis[0].strip())
            BOOK.append(lis[1].strip())
            author = lis[2].strip() if lis[2].strip() != '' else ''
            AUTHOR.append(author)
    return ID, BOOK, AUTHOR


book_id, book_name, author = get_books(filepath)
book_info = tuple(zip(book_name, author))
book_info = dict(zip(book_id, book_info))
# print(book_info)
```

```python
# ID    |READER|MINUTES| SUBSET           | PROJ.|BOOK ID| CH. TITLE | PROJECT TITLE
filepath = '../data/LibriSpeech/CHAPTERS.TXT'

def get_chapters(filepath):
    ID = []
    READER = []
    MINUTES = []
    SUBSET = []
    PROJ = []
    BOOK = []
    TITLE = []
    PROJECT_TITLE = []
    with open(filepath, 'r') as f:
        lines = f.readlines()
        for line in lines:
            # print(line)
            lis = str(line).split('|')
            # print(lis)
            ID.append(lis[0].strip())
            READER.append(lis[1].strip())
            MINUTES.append(lis[2].strip())
            SUBSET.append(lis[3].strip())
            PROJ.append(lis[4].strip())
            BOOK.append(lis[5].strip())
            TITLE.append(lis[6].strip())
            PROJECT_TITLE.append(lis[7].strip())
    return ID, READER, MINUTES, SUBSET, PROJ, BOOK, TITLE, PROJECT_TITLE


chapter_id, reader, _, _, _, chapter_book, chapter_title, _ = get_chapters(filepath)
# print(reader)
```

```python
for i in range(len(chapter_id)):
    if reader[i] in data:
        idx = chapter_id[i]
        if chapter_book[i] in book_info:
            book = book_info[chapter_book[i]][0]
            author = book_info[chapter_book[i]][1]
        else:
            book = ''
            author = ''
        data[reader[i]][idx] = {
            'book': book,
            'author': author,
            'title': chapter_title[i]
        }
        # data[reader[i]]['chapter']['id'] = idx
        # data[reader[i]]['chapter']['book'] = book_info[chapter_book[i]][0]
        # data[reader[i]]['chapter']['author'] = book_info[chapter_book[i]][1]
        # data[reader[i]]['chapter']['title'] = chapter_title[i]

print(data)
json_data = json.dumps(data, indent=4)
print(json_data)
```
